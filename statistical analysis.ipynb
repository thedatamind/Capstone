```python
# CHI-SQUARE TEST FRO INDEPENDANCE # pages 6, 7 and 8 of the project

# https://docs.scipy.org/doc/scipy-1.7.0/reference/reference/generated/scipy.stats.chi2_contingency.html#:~:text=This%20is%20a%20test%20for,square%20statistic%20equal%20to%200.
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, chi2_contingency

# The contingency table
observed = np.array([
    [28, 40],
    [91, 82],
    [32, 45],
    [212, 124],
    [59, 26],
    [169, 86],
    [376, 156],
    [46, 34]
])

# Perform the chi-squared test
# https://docs.scipy.org/doc/scipy-1.7.0/reference/reference/generated/scipy.stats.chi2_contingency.html#:~:text=This%20is%20a%20test%20for,square%20statistic%20equal%20to%200.
dof = 7  # Degrees of freedom
chi2_statistic, p, dof, expected = chi2_contingency(observed)

# Degrees of freedom
dof = observed.size - sum(observed.shape) + observed.ndim - 1

# Significance level (alpha)
alpha = 0.05

# Critical chi-squared value
# https://docs.scipy.org/doc/scipy-1.7.0/reference/reference/generated/scipy.stats.chi2_contingency.html#:~:text=This%20is%20a%20test%20for,square%20statistic%20equal%20to%200.
critical_chi2 = chi2.ppf(1 - alpha, dof)

# Plot the curve with x-values up to 300
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
x = np.linspace(0, 300, 1000)  # Set x-values up to 300
plt.plot(x, chi2.pdf(x, dof), label=f'Chi-squared Distribution (df={dof})')
plt.axvline(x=chi2_statistic, color='r', linestyle='--', label='Observed Chi-squared')
plt.axvline(x=critical_chi2, color='g', linestyle='--', label=f'Critical Chi-squared (α={alpha})')
plt.fill_between(x, 0, chi2.pdf(x, dof), where=(x >= critical_chi2), color='green', alpha=0.3)
plt.fill_between(x, 0, chi2.pdf(x, dof), where=(x <= chi2_statistic), color='red', alpha=0.3)
plt.xlabel('Chi-squared Value')
plt.ylabel('Density')
plt.title('Chi-squared Test')
plt.legend()
plt.show()

# Print results
print("Chi-squared statistic:", chi2_statistic)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)
print("Critical chi-squared value (alpha=0.05):", critical_chi2)
```

```python
# This is the program used to scramble the emails
# https://www.geeksforgeeks.org/md5-hash-python/

import pandas as pd
import hashlib

# Path to the data
file_path = r'C:\Users\Ilan\Desktop\Capstone\Capstone Data\Combined data renamed.xlsx'
df = pd.read_excel(file_path)

# A for scrambling the email addresses
def scramble_email(email):
    # Check if the email is NaN or missing
    if pd.isna(email):
        return email
    # Convert to lowercase 
    email_lower = email.lower()
    # https://www.geeksforgeeks.org/md5-hash-python/
    hashed_email = hashlib.md5(email_lower.encode()).hexdigest()
    return hashed_email

# Scramble the emails
scrambled_emails = df['Customer Email'].apply(scramble_email)

# Replacinf the velues with the new ones
df['Customer Email'] = scrambled_emails

# Saving and exporting
df.to_excel(file_path, index=False)

```
